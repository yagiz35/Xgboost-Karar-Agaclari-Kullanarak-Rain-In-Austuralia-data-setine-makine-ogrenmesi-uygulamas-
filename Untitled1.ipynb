## Kütüphaneler dahil edildi ve Veri Seti Tanımlandı

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("weatherAUS.csv")
```

## Veri setinde bulunan date(tarih) kategorisi object olarak bulunmakta, bu kategori 

```python
data["Date"] = data["Date"].apply(pd.to_datetime)
```

```python
data.head()
```

### Veri Setinin incelenmesi sonucu, kategorilerin Kategorik ve Numerik olmak üzere çeşitli tipte verilere sahip olduğunu ve bir takım eksik veri barındırdığı sonucunu görüyoruz

- Eksik verilerle başa çıkmak adına veri setini yeniden indexleyerek, kategorik ve numerik veriler ayrıldı.

```python
data = data.reindex(["Date","MinTemp","MaxTemp","Rainfall","Evaporation",
            "Sunshine","WindGustSpeed","Humidity9am",
            "Humidity3pm","Pressure9am","Pressure3pm",
            "Cloud9am","Cloud3pm","Temp9am","Temp3pm","Location","WindGustDir","WindDir9am","RainToday","RainTomorrow"], axis=1)
```

```python
data.head()
```

```python
categorical = [x for x in data.columns[15:-1]]
```

```python
plt.figure(figsize = (18,18))
for i,c in enumerate(categorical):
    plt.subplot(2,2,i+1)
    sns.countplot(data[c], hue = data["RainTomorrow"])
    plt.title(c)
    plt.xlabel("")
```

```python
numvalues = [clmn for clmn in data.columns[1:15]]
```

data.head()

```python
data[numvalues].hist(figsize=(16, 20), bins=50, xlabelsize=8, ylabelsize=8)
```

```python
data.isnull().sum()
```

## Kategorik verilerde mod, numerik verilerde medyan 

```python
data.info()
```

```python
plt.figure(figsize=(15,13))
data.boxplot(numvalues, rot=30)
```

```python
sayac = 0
for feature in numvalues:
    sayac = sayac + 1
    q1 = data[feature].quantile(0.25)
    q3 = data[feature].quantile(0.75)
    IQR = q3-q1
    lower_limit = q1 - (IQR*1.5)
    upper_limit = q3 + (IQR*1.5)
    data.loc[data[feature]<lower_limit, feature] = lower_limit
    data.loc[data[feature]>upper_limit, feature] = upper_limit
        
print(f"{sayac} kadar veri cikarildi")
```

```python
plt.figure(figsize=(15,13))
data.boxplot(numvalues, rot=30)
```

```python
for column in data.columns[1:15]:
    data[column] = data[column].fillna(data[column].mean())
```

```python
for column in data.columns[15:]:
    data[column] = data[column].fillna(data[column].mode()[0])
```

```python
data.isnull().sum()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
for column in data.columns[15:]:
    data[column] = le.fit_transform(data[column])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop(["RainTomorrow","Date"],axis = 1)
Y = data["RainTomorrow"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2)
```

```python
y_train.value_counts()
```

```python
plt.figure(figsize =(20,9))
sns.countplot(y_train)
plt.show()
```

```python
plt.figure(figsize =(20,9))
sns.countplot(y_test)
plt.show()
```

```python
y_test.value_counts()
```

```python
from sklearn.linear_model import LogisticRegression
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import StandardScaler
```

```python
smote_over_sampling = SMOTE(random_state=1 ,sampling_strategy="auto")

x_train_resampled, y_train_resampled = smote_over_sampling.fit_resample(X_train, y_train)
```

```python
scaler = StandardScaler()

x_train_resampled = scaler.fit_transform(x_train_resampled)

X_test = scaler.transform(X_test)
```

```python
logreg = LogisticRegression(solver='liblinear', random_state=0)
logreg.fit(x_train_resampled, y_train_resampled)
```

```python
logreg.fit(x_train_resampled, y_train_resampled)
```

```python
preds = logreg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(y_test, preds)
```

```python
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(logreg, X_test,y_test)
```

```python
print('Training set score: {:.4f}'.format(logreg.score(X_train, y_train)))

print('Test set score: {:.4f}'.format(logreg.score(X_test, y_test)))
```

```python
importance = logreg.feature_importances_
# summarize feature importance
for i,v in enumerate(importance):
    print('Feature: %0d, Score: %.5f' % (i,v))
# plot feature importance
sns.barplot([x for x in range(len(importance))], importance)
```
