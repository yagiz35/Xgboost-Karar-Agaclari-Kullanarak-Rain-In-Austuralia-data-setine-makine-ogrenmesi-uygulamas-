```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Kütüphaneler eklendi

```python
data = pd.read_csv("weatherAUS.csv")
```

```python
data.describe()
```

```python
data.dropna(inplace = True)
```

```python
data.sort_values(by="Date",inplace = True)
```

## [-6.7] cıkarılamadi

```python
for column in data.columns[2:-1]:
    sns.scatterplot(data = data , x="Date",y=column,hue="RainTomorrow")
    plt.show()
```

```python
data.columns[2:-2]
```

```python
data = data.reindex(["Date","Location","MinTemp","MaxTemp","Rainfall","Evaporation",
            "Sunshine","WindGustSpeed","Humidity9am",
            "Humidity3pm","Pressure9am","Pressure3pm",
            "Cloud9am","Cloud3pm","Temp9am","Temp3pm","WindGustDir","WindDir9am","RainToday","RainTomorrow"], axis=1)
```

```python
data
```

```python
for column in data.columns[2:-5]:
    for ifrainy in data["RainTomorrow"].unique():
        selected_ifrainy = data[data["RainTomorrow"] == ifrainy]
        selected_column = selected_ifrainy[column]
        
        std = selected_column.std()
        avg = selected_column.mean()
        
        three_sigma_plus = avg + (3 * std)
        three_sigma_minus = avg - (3* std)

        outliers = selected_column[(selected_ifrainy[column] > (three_sigma_plus)) | (selected_ifrainy[column] < (three_sigma_minus))].index
        data.drop(index = outliers ,inplace = True)                            
        
        
    
```
